```
import torch

from pytorch_pretrained_bert import BertTokenizer

tokenizer = BertTokenizer.from_pretrained('bert-base-uncased')
```

## Insepting Vocabulary

### 1. Vocab Dump  

Retrive the list of "tokens" in BERT vocabulary into a text file

```
with open("bert_vocab.txt", 'w') as f:

    for token in tokenizer.vocab.keys():
        f.write(token + '\n')
```

### 2. Checking Single Characters

```
one_chars = []
one_chars_hashes = []

for token in tokenizer.vocab.keys():
    if len(token) == 1:
        one_chars.append(token)
    
    elif len(token) == 3 and token[0:2] == "##":
        one_chars_hashes.append(token[2])
```

```
print("Number of single Character tokens", len(one_chars), '\n')

##Printing all single character 40 per row
for i in range(0, len(one_chars), 40):
    end = min(i + 40, len(one_chars)+1)
    ##print(end,'\n')
    print(' '.join(one_chars[i:end])) 
```

```
print("Number of Single Charcaters with hashes", len(one_chars_hashes), '\n')

for i in range(0, len(one_chars_hashes), 40):
    end = min(i+40,len(one_chars_hashes)+1)
    print(" ".join(one_chars_hashes[i:end]))
```

```
identical = set(one_chars) == set(one_chars_hashes)

print("Are the sets identical ------->", identical)
```

### 3. Subwords vs Whole Words

```
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
%matplotlib inline
sns.set(style = 'darkgrid')
sns.set(font_scale = 1.5)
plt.rcParams["figure.figsize"] = (16,9)
```

```
token_length = [len(token) for token in tokenizer.vocab.keys()]

sns.countplot(token_length)
plt.title('Token/Word length in Vocabulary')
plt.xlabel('Token Length')
plt.ylabel('Number of Token')


print("Maximum token length ------>", max(token_length))
```

```
##For all tokens starting with '##'

all_chars_hashes = []

for token in tokenizer.vocab.keys():
    if len(token) >= 2 and token[0:2] == "##":
        all_chars_hashes.append(token[2:])

print("Number of all subwords in vocabluary ----->", len(all_chars_hashes))
```

```
##Percentage of all tokens in vocabluary

percentage = len(all_chars_hashes)/len(tokenizer.vocab.keys()) 
print("Total number of Tokens ---->",len(tokenizer.vocab.keys()), "\nTotal number of Subwords---->", len(all_chars_hashes))
print('\nPercentage of Subwords----->',percentage*100,"%")
```

```
subword_length = [len(token) for token in all_chars_hashes]


sns.countplot(subword_length)
plt.title('SubWord Token length without ##')
plt.xlabel('Token Length')
plt.ylabel('Number of Token')


print("Maximum token length(excluding '##') ------>", max(subword_length))
```

## 4. Checking Names in Vocabulary

```
!pip install wget
```

```
import wget
import random

url = 'http://www.gutenberg.org/files/3201/files/NAMES.TXT'

wget.download(url, 'first_name_list.txt')
```

```
##Reading the names and saving them in lower case
with open('first_name_list.txt', 'rb') as f:
    names_encoded = f.readlines()

names = []

for name in names_encoded:
    try:
        names.append(name.rstrip().lower().decode('utf-8'))
    except:
        continue

print('Number of names: {:,}'.format(len(names)))
print('Example: ', random.choice(names))
```

```
num_names = 0

# For each name in our list...
for name in names:

    # If it's in the vocab...
    if name in tokenizer.vocab:
        # Tally it.
        num_names += 1

print('{:,} names in the vocabulary'.format(num_names))
```

## 5. Numbers in Vocabluary

```
# Count how many numbers are in the vocabulary.
count = 0

# For each token in the vocabulary...
for token in tokenizer.vocab.keys():

    # Tally if it's a number.
    if token.isdigit():
        count += 1
        
        # Any numbers >= 10,000?
        if len(token) > 4:
            print(token)

print('Vocab includes {:,} numbers.'.format(count))
```

```
# Count how many dates between 1600 and 2021 are included.
count = 0 
for i in range(1600, 2021):
    if str(i) in tokenizer.vocab:
        count += 1

print('Vocab includes {:,} of 421 dates from 1600 - 2021'.format(count))
```
